```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
data= pd.read_csv("Japan_life_expectancy.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(['Prefecture'],axis=1) 
```

```python
data['Life_expectancy'] = data['Life_expectancy'].astype(int)
```

```python
data['Welfare_exp'] = data['Welfare_exp'].astype(int)
```

```python
display(data.dtypes)
```

```python
life_expectancy= data[["Life_expectancy"]]
welfare_exp= data[["Welfare_exp"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(life_expectancy,welfare_exp,test_size=1/3,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
lin= LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
print(lin.intercept_)
```

```python
print(lin.coef_)
```

```python
pred=lin.predict(x_test)
```

```python
print(pred)
```

```python
from sklearn import metrics
```

```python
error=np.sqrt(metrics.mean_squared_error(y_test,pred))
```

```python
print(f"THE ACCURACY IS ={abs(error-100)}")
```
